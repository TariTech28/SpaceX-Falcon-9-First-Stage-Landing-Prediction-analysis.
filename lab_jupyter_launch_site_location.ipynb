```python
# TASK 3: Calculate the distances between a launch site to its proximities
from geopy.distance import distance
launch_sites_df
coast_point = [28.56367, -80.57163]  # near CCAFS coast
for index, row in launch_sites_df.iterrows():
    
    site_coord = (row['Lat'], row['Long'])
    
    dist = distance(site_coord, coast_point).km
    
    print(f"{row['Launch Site']} is {dist:.2f} km from coastline")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

# Findings from Launch Site Proximity Analysis
🚀 1. Are launch sites in close proximity to railways?

👉 Partially, yes

Several SpaceX launch sites are located within reasonable distance of railway infrastructure.
However, they are not directly adjacent to rail lines.
This reduces vibration risk and avoids interference with launch operations.

✔ Conclusion:

Launch sites are moderately close but intentionally not directly on railways.
🛣️ 2. Are launch sites in close proximity to highways?

👉 Yes

Most launch sites (especially in Florida) are close to major highways.
Example: Kennedy Space Center area has highway access nearby.

✔ Reason:

Easy transport of rockets, equipment, and personnel
Efficient logistics without being inside urban congestion

✔ Conclusion:

Launch sites are close to highways for logistics convenience.

🌊 3. Are launch sites in close proximity to coastline?

👉 Yes — very close (key design choice)

All major SpaceX launch sites are within a few kilometers of the coast.
Some are almost directly on coastal edges.

✔ Reason:

Rockets launch over the ocean (eastward)
Prevents debris from falling on populated land
Safety + regulatory compliance

✔ Conclusion:

Launch sites are intentionally placed very close to coastlines.

🏙️ 4. Do launch sites keep distance away from cities?

👉 Yes — clearly

Sites are generally far from dense urban centers.
Example:
Orlando is tens of kilometers away from Cape Canaveral sites.
California site (Vandenberg) is also away from major cities.

✔ Reason:

Safety buffer zone for launches
Noise and explosion risk control
Regulatory requirements

✔ Conclusion:

Launch sites are strategically distant from cities.
